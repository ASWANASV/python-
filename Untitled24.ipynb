# Group the data by team

```python
team_counts = df['Team'].value_counts()
```

```python
print(df.groupby('Team').size())
```

# Calculate the percentage split of employees in each team

```python
team_percentages = team_counts / len(df) * 100
print(team_percentages)
```

```python
plt.figure(figsize=(8, 6))
plt.pie(team_percentages, labels=team_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Employees Across Teams')
plt.show()
```

```python
position_counts = df.groupby('Position').size()
print(position_counts)
```

```python
age_counts = df['Age'].value_counts()
print(age_counts)
```

```python
salary_expenditure = df.groupby(['Team', 'Position'])['Salary'].sum()
print(salary_expenditure)
```

```python
correlation = df['Age'].corr(df['Salary'])
print(correlation)
```

```python
import matplotlib.pyplot as plt

team_counts.plot(kind='bar')
plt.title('Distribution of Employees Across Teams')
plt.xlabel('Team')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt

position_counts = df['Position'].value_counts()
position_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribution of Employees by Position')
plt.show()
```

```python
import matplotlib.pyplot as plt

df['Age'].plot(kind='hist', bins=10)
plt.title('Distribution of Employees by Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt

salary_expenditure = df.groupby(['Team', 'Position'])['Salary'].sum()
salary_expenditure.plot(kind='bar')
plt.title('Total Salary Expenditure by Team and Position')
plt.xlabel('Team and Position')
plt.ylabel('Total Salary Expenditure')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Age'], df['Salary'])
plt.title('Correlation Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```
